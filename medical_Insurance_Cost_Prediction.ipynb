```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("/content/sample_data/insurance.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.groupby('region').size()
```

```python
sexq={'male':1,'female':2}
reg={'northeast':1,'northwest':2,'southeast':3,'southwest':4}
somke={"yes":1,"no":0}
```

```python
data['sex']=data['sex'].map(sexq)
data['region']=data['region'].map(reg)
data['smoker']=data['smoker'].map(somke)
data
```

```python
import pickle
datast=pickle.dump(data,open("medical Insurance Cost Prediction.csv",'wb'))
```

```python
pickle.load(open("medical Insurance Cost Prediction.csv",'rb'))
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]

```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=20)
model=LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
r2=r2_score(y_pred,Y_test)
print(r2)
```

```python
import pickle
model=pickle.dump(model,open("medical Insurance Cost Prediction.pkl",'wb'))
```
